```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

```python
data=pd.read_csv('/content/gld_price_data.csv')
```

```python
correlation=data.drop('Date',axis=1).corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```

```python
print(correlation['GLD'])
```

```python
#checking for distribution of the GLD Price
sns.distplot(data['GLD'],color='g')
```

```python
# splitting the data
x=data.drop(['GLD','Date'],axis=1)
y=data['GLD']
```

```python
print(x,y)
```

```python
# splitting the training and testing data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
# model training
model=RandomForestRegressor(n_estimators=100)
model.fit(x_train,y_train)
```

```python
# model evaluation
x_train_prediction=model.predict(x_train)
error_score=metrics.r2_score(y_train,x_train_prediction)
print('R squared error : ',error_score)
```

```python
plt.scatter(y_train,x_train_prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Price vs Predicted Price')
plt.show()
```

```python
x_test_prediction=model.predict(x_test)
error_score=metrics.r2_score(y_test,x_test_prediction)
print('R squared error : ',error_score)
```

```python
plt.scatter(y_test,x_test_prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Price vs Predicted Price')
plt.show()
```

```python
Y_test=list(y_test)
```

```python
plt.plot(Y_test,color='blue',label='Actual Price')
plt.plot(x_test_prediction,color='green',label='Predicted Price')
plt.title('Actual Price vs Predicted Price')
plt.xlabel('Number of values')
plt.ylabel('GLD Price')
plt.legend()
plt.show()
```
